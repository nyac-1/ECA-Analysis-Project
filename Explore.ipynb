```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
googleOne = pd.read_csv("./g/googleplaystore.csv")
googleOne.fillna(0.0,inplace = True)
appleOne = pd.read_csv("./a/AppleStore.csv")
```

```python
#Now try to streamline both datasets.
#first change all the object values to numeric values in the google data set and rename them
```

```python
remove_google_columns = ['Genres','Last Updated','Current Ver', 'Android Ver','Installs','Type']
remove_apple_columns = ['Unnamed: 0','currency','id','rating_count_ver','user_rating_ver','ver','sup_devices.num','ipadSc_urls.num','lang.num','vpp_lic']
```

```python
googleOne.drop(remove_google_columns, axis = 1, inplace = True)
appleOne.drop(remove_apple_columns, axis = 1, inplace = True)
```

```python
googleOne.head()
```

```python
appleOne.head()
```

```python
google_rename = {"App":"name","Category":"category", "Rating":"rating",'Reviews':'nreviews','Size':'size','Price':'price','Content Rating':'content_rating'}
apple_rename = {"track_name":"name","prime_genre":"category","user_rating":"rating",'cont_rating':'content_rating','size_bytes':'size'}

googleOne.rename(columns=google_rename, inplace = True)
appleOne.rename(columns=apple_rename, inplace = True)
```

```python
googleOne.to_csv("main_google.csv", index = False)
appleOne.to_csv("main_apple.csv", index = False)
```
